# Data Visualization with seaborn

```python
# import the required libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline

# set the style of the plots
sns.set_style("whitegrid")

print("Setup Complete")
```

```python
# import the data to be used
df = pd.read_csv("../data/WorldCupMatchesCleaned.csv", parse_dates=True, index_col="Year")

# check the data was loded correctly
df.head()
```

```python
# create a figure with bigger size
plt.figure(figsize=(16,6))

# set title of the plot
plt.title("Attendance in World Cup matches")

# plot the attendance column with a line plot from seaborn
sns.lineplot(data=df['Attendance'])
```

```python
# create a figure with bigger size
plt.figure(figsize=(12,6))

# set title of the plot
plt.title("Attendance in World Cup matches")

# bar plot of the attendance column
index = pd.to_datetime(df.index)
sns.barplot(y=df['Attendance'], x=index.year)
```

```python
# get the total goals scored in each match
df['GoalsScored'] = df['Home Team Goals'] + df['Away Team Goals']
```

```python
# plot the relation between attendance and goals scored
sns.scatterplot(x=df['GoalsScored'], y=df['Attendance'])
```

```python
# see regression line
sns.regplot(x=df['GoalsScored'], y=df['Attendance'])
```

```python
# create a figure with bigger size
plt.figure(figsize=(18,9))

# for catigorical data use this scatter plot
sns.swarmplot(x=df['GoalsScored'], y=df['Attendance'])
```

```python
# histogram of the attendance column
sns.histplot(df['Attendance'])
```

```python
# KDE plot 
sns.kdeplot(data=df['Attendance'], shade=True)
```
